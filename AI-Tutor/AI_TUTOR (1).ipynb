### Steps to set `OPENAI_API_KEY` in Google Colab Secrets:

1.  **Open the Secrets tab**: On the left-hand side of your Colab notebook, look for the '🔑' (key) icon in the vertical toolbar. Click on it.
2.  **Add a new secret**: Click on the 'Add new secret' button.
3.  **Name the secret**: In the 'Name' field, type `OPENAI_API_KEY` (it must be exactly this name for the code to find it).
4.  **Enter your API Key**: In the 'Value' field, paste your actual OpenAI API key.
5.  **Enable notebook access**: Make sure the 'Notebook access' toggle is switched ON for the `OPENAI_API_KEY` secret.
6.  **Run the following code cell** to retrieve and set the API key from the secrets manager.

After running the above cell, you should be able to use the `openai` library without encountering the API key error.
