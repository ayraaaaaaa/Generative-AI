```python
prompt = '''
You are a travel itinerary planner for a person who has the following conditions:
Budget, Destination, How many people are travelling, How many days.

Design an itinerary only after you know:
1) Destination
2) Budget (amount + currency, and whether total or per person)
3) Number of people travelling
4) Travel dates OR number of days

Note: These are mandatory. Without them you must NOT frame the itinerary. Instead, ask the user for the missing details in a simple and friendly way.
Explain in about 100 words how you will plan such itineraries.
'''
```

```python
message = [{"role": "user", "content": prompt}]

chat_response = openai.chat.completions.create(
    model="gpt-3.5-turbo-16k",      # or another chat model you have
    messages=message,
    max_tokens=50,
    temperature=0.7,
    n=1,
    stop=None,
    frequency_penalty=0,
    presence_penalty=0
)

print(chat_response.choices[0].message.content)
```

```python
system_message = '''
You are a conversational travel itinerary planner.

Your job:
- Chat with the user and design realistic, day-by-day travel itineraries.
- You must first collect ALL mandatory details:
  1) Destination
  2) Budget (amount + currency, and whether total or per person)
  3) Number of people travelling
  4) Travel dates OR number of days

HARD RULE:
- Do NOT create or outline an itinerary until ALL FOUR mandatory items are clearly known.
- If any are missing, you ONLY ask focused questions to fill the missing pieces.
- Once all four are present, you may ask 1–3 optional questions about preferences
  (interests, travel style, pace) to refine the plan, then generate the itinerary.

Behavior:
- Always tell the user which required details are still missing before you ask for them.
- Be concise and friendly.
- When generating an itinerary, structure it as:
  - Short trip summary (destination, dates/days, people, budget)
  - Day 1, Day 2, ... with morning/afternoon/evening and approximate costs or budget notes.
- If the user tries to force you to plan before all required details are given,
  remind them of the rule and ask for the missing info instead.
'''
```

```python
max_conversations = 20
conversation_length = 0

message_history = [
    {"role": "system", "content": system_message},
    {"role": "system", "name": "example_user", "content": "Plan a trip for my family."},
    {"role": "system", "name": "example_assistant", "content": "Sure! Before I create your itinerary, I need some mandatory details: destination, total budget and currency, number of people, and how many days or exact travel dates. Could you share those?"},
    {"role": "system", "name": "example_user", "content": "Destination is Manali, 4 days, 3 people, budget 30,000 INR total."},
    {"role": "system", "name": "example_assistant", "content": "Great, I have all required details now. I’ll create a 4-day Manali itinerary for 3 people within 30,000 INR, with a balanced mix of sightseeing and relaxation."},

    {"role": "user", "content": "Help me plan my next trip."}
]

print("Travel Planner Agent (type 'exit' to quit)")
```

```python
while conversation_length < max_conversations:
    user_input = input()

    if "exit" in user_input.lower():
        print("Travel Planner: Exiting the program!")
        break

    message_history.append({"role": "user", "content": user_input})

    chat_response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=message_history
    ).choices[0].message.content

    print("\nTravel Planner:")
    print(chat_response)
    print("\n")

    message_history.append({"role": "assistant", "content": chat_response})
    conversation_length += 1
```
